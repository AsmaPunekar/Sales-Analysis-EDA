```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the figure
plt.figure(figsize=(12, 6))

# Box plot for Sales and Unit
sns.boxplot(data=df[['Sales', 'Unit']], palette='pastel')


plt.title('Box Plot for Sales and Unit with outliers')
plt.ylabel('Values')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
def count_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return ((series < lower) | (series > upper)).sum()
```

```python
# Extract the Sales column

sales_outliers = count_outliers(df['Sales'])
# Summary output
print("Outlier Analysis for Overall Sales")
print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR: {IQR}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Number of Outliers: {(sales_outliers)}")
```

```python
# Extract Unit column
units_outliers = count_outliers(df['Unit'])
# Summary output
print("Outlier Analysis for Units")
print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR: {IQR}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Number of Outliers: {(units_outliers)}")
```

### Conclusion
-  Both Unit and sales feature are rigth skewed and having outliers also
-  Therefore, using IQR method, it is proved that there are 123 number of outliers.

```python
df['Date'] = pd.to_datetime(df['Date'])

# Extract date parts
df['Day'] = df['Date'].dt.date
df['Week'] = df['Date'].dt.isocalendar().week
df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = df['Date'].dt.quarter
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.set(style='whitegrid')

# Create 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(16, 10))

# --- Daily Sales ---
daily_sales = df.groupby('Day')['Sales'].sum().reset_index()
sns.lineplot(data=daily_sales, x='Day', y='Sales', marker='o', ax=axs[0, 0])
axs[0, 0].set_title('Daily Sales Trend')
axs[0, 0].set_xlabel('Date')
axs[0, 0].set_ylabel('Total Sales')
axs[0, 0].tick_params(axis='x', rotation=45)

# --- Weekly Sales ---
weekly_sales = df.groupby('Week')['Sales'].sum().reset_index()
sns.barplot(data=weekly_sales, x='Week', y='Sales', palette='Blues',hue="Week", ax=axs[0, 1],legend = False)
axs[0, 1].set_title('Weekly Sales Trend')
axs[0, 1].set_xlabel('Week Number')
axs[0, 1].set_ylabel('Total Sales')

# --- Monthly Sales ---
ordered_months = ['October', 'November', 'December']
monthly_sales = df.groupby('Month')['Sales'].sum().reindex(ordered_months).reset_index()
sns.barplot(data=monthly_sales, x='Month', y='Sales', palette='Oranges',hue="Month", ax=axs[1, 0],legend= False)
axs[1, 0].set_title('Monthly Sales Trend')
axs[1, 0].set_xlabel('Month')
axs[1, 0].set_ylabel('Total Sales')

# --- Quarterly Sales ---
quarterly_sales = df.groupby('Quarter')['Sales'].sum().reset_index()
sns.barplot(data=quarterly_sales, x='Quarter', y='Sales',palette='Purples',ax=axs[1, 1],legend =False)
axs[1, 1].set_title('Quarterly Sales Trend')
axs[1, 1].set_xlabel('Quarter')
axs[1, 1].set_ylabel('Total Sales')

# Adjust layout
plt.tight_layout()
plt.show()
```

### Visualization tools
- Why Seaborn, Matplotlib, and Plotly are Used:
  - Seaborn:
    Chosen for its powerful statistical plotting capabilities and visually appealing default themes. It is ideal for box plots, bar plots, and distribution plots,        especially for analyzing sales trends across groups and time periods.

  - Matplotlib:
    Used alongside Seaborn for fine-tuned control over plot formatting, layout, and customizations. It is especially useful for subplots, axis formatting, and adding     labels/titles when Seaborn alone is not sufficient.

  - Plotly:
    Selected for building interactive visualizations like sunburst charts, which allow dynamic exploration of multi-level data (e.g., Sales by State → Group). It         enhances user experience, especially for dashboards intended for stakeholders or presentations.
